```python
import matplotlib.pyplot as plt
import numpy as np
from astropy.io import fits
from astropy.wcs import WCS
from pv_maker import plot_pv
import astropy.units as u
```

```python
basepath = '/orange/adamginsburg/ACES/broadline_sources/EVFs/images/'
```

```python
pos = 0.23
pv_fn = f'{basepath}/CS21_pv_b{pos}_max.fits'
pv1 = fits.open(pv_fn)
plot_pv(pv1[0], 'CS21', pos*u.deg, longitude=False, mean='max', close=False)
```

```python
pos = -0.482
pv_fn = f'{basepath}/CS21_pv_l{pos}_max.fits'
pv1 = fits.open(pv_fn)
plot_pv(pv1[0], 'CS21', pos*u.deg, longitude=True, mean='max', close=False)
```

```python
from pv_maker import make_position_list
```

```python
b_min = -0.27*u.deg
b_max = 0.22*u.deg
list_b = make_position_list(b_min, b_max)
#list_b = list_b.to(u.deg)

l_min = -0.59*u.deg
l_max = 0.88*u.deg
list_l = make_position_list(l_min, l_max)
#list_l = list_l.to(u.deg)
```

```python
for pos in list_b:
    pv_fn = f'{basepath}/CS21_pv_b{round(pos.value, 3)}_max.fits'
    pv1 = fits.open(pv_fn)
    plot_pv(pv1[0], 'CS21', pos, longitude=False, mean='max', close=True)
```

```python
for pos in list_l:
    pv_fn = f'{basepath}/CS21_pv_l{round(pos.value, 3)}_max.fits'
    pv1 = fits.open(pv_fn)
    plot_pv(pv1[0], 'CS21', pos, longitude=True, mean='max', close=True)
```
